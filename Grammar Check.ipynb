```python
import language_check
import os
```

```python
tool = language_check.LanguageTool('en-US')
```

```python
# Can be replaced by other APIs
def grammar_check(text):
    matches = tool.check(text)
    count = 0
    for match in matches:
        if match.ruleId != 'MORFOLOGIK_RULE_EN_US' and match.ruleId != 'EN_A_VS_AN' and match.ruleId != 'I_LOWERCASE':
#             print(match)
            count += 1
    return count

# Parse the text into a string
def parse_text(text):
    text = list(filter(lambda x: x != '\n', text))
    text = ' '.join(text)
    text = text.replace(' ,', ',')
    text = text.replace(" '", "'")
    text = '. '.join(list(map(lambda x: x.strip().capitalize(), text.split('.'))))
    return text
```

```python
import pickle

poem = pickle.load(open("Feb_DTS_no_story/alien_female_25_30_0.1_multi_True_no_story.pickle", "rb" ))
```

```python
text = poem[50][2]
print(grammar_check(parse_text(text)))
```

```python
good_poems = []
def print_all_good_poems(directory):
    for file in os.listdir(directory):
        if file.endswith(".pickle"):
            poem = pickle.load(open(os.path.join(directory, file), "rb"))
            try:
                if len(poem) >= 1:
                    text = parse_text(poem[0][2])
                    if grammar_check(text) == 0:
                        good_poems.append(' '.join(poem[0][2]))
                        print(' '.join(poem[0][2]))
            except:
                continue
    print(len(good_poems))
```

```python
print_all_good_poems("Feb_DTS_story")
```
