```python

from py_files.Limericks import Limerick_Generate


# In[4]:
```

```python

lg = Limerick_Generate(model_dir='/gpt2/models/345M',model_name='345M')

```

```python

orig_prompt = """Love is too young to know what conscience is yet who knows not conscience is born of love."""
prompt = lg.enc.encode(orig_prompt)
#prompt = lg.enc.encode("""Love conscience is conscience is love""") 
#prompt = lg.enc.encode("""Truth""")


default_templates=[['NNP', 'VBZ', 'RB', 'JJ', 'TO', 'VB', 'WP', 'NN', 'VBZ', 'CC', 'WP', 'VBZ', 'RB', 'NN', 'VBZ', 'VBN', 'IN', 'NN']]

rep_templates=[[11, 21, 0, 0, 0, 0, 0, 31, 22, 0, 0, 0, 0, 32, 23, 0, 0, 12]]

#prompt = lg.enc.encode("""Love's fire heats water water cools not love.""")
#default_templates=[['NNP','POS','NN','VBZ','NN','NN','VBZ','RB','VB']]
#cooccurance_templates=[[11,0,0,0,21,22,0,0,12]]

banned_set = ['love', 'conscience']
new_sentences = lg.gen_line_gpt_cc(cctemplate=rep_templates[0], w=None, encodes=prompt, default_template =
default_templates[0], banned_set=banned_set,search_space=100)

sents = [new_sentences[j][0] for j in range(100)]
sent_done = [' '.join(sents[j]) for j in range(100)]
print(sent_done)
```

```python
f = open('output1.txt', 'w')
f.write(" \n".join(sent_done))
f.close()
```
